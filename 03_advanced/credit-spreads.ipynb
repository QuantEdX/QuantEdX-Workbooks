## 1. Introduction

Credit spreads compensate investors for expected loss, liquidity risk, and uncertainty. Spread analysis should separate changes in benchmark rates from changes in issuer-specific credit conditions.


```python
bonds = pd.DataFrame({'issuer': ['A', 'B', 'C'], 'pd': [0.01, 0.025, 0.05], 'lgd': [0.45, 0.50, 0.60], 'risk_premium': [0.008, 0.012, 0.020]})
bonds['spread_estimate'] = bonds['pd'] * bonds['lgd'] + bonds['risk_premium']
bonds
```

## 2. Mathematical Foundation

A simple approximation decomposes a credit spread into expected loss and a risk premium:

$$
s \approx \operatorname{PD} \times \operatorname{LGD} + \text{risk premium}.
$$
