## 1. Introduction

Graph models represent entities as nodes and relationships as edges. In finance, graphs can describe payments, ownership links, transaction networks, or counterparty exposures.


```python
edges = pd.DataFrame({'source': ['A', 'A', 'B', 'C'], 'target': ['B', 'C', 'C', 'D']})
degree = pd.concat([edges['source'], edges['target']]).value_counts()
degree.rename('degree_centrality')
```

## 2. Mathematical Foundation

For adjacency matrix \(A\), degree centrality of node \(i\) is:

$$
C_D(i)=\sum_j A_{ij}.
$$

High centrality may indicate influence, concentration, or systemic importance.
