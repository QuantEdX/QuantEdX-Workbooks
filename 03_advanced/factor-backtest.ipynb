## 1. Introduction

A factor backtest ranks assets using lagged information, forms portfolios at scheduled rebalances, and measures returns after realistic trading costs. Point-in-time data is essential.


```python
returns = pd.DataFrame({'long_leg': [0.012, -0.004, 0.009], 'short_leg': [0.003, 0.002, -0.005], 'cost': [0.001, 0.001, 0.001]})
returns['factor_return'] = returns['long_leg'] - returns['short_leg'] - returns['cost']
returns
```

## 2. Mathematical Foundation

For long and short portfolio returns \(r_{L,t}\) and \(r_{S,t}\):

$$
r_{\text{factor},t}=r_{L,t}-r_{S,t}-\text{cost}_t.
