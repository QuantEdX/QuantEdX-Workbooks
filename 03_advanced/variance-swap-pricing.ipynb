## 1. Introduction

A variance swap pays the difference between realized variance and a strike fixed at inception. Because payoff is linear in variance rather than volatility, large market moves matter disproportionately.


```python
variance_notional = 1000
realized_vol = 0.24
variance_strike = 0.20**2
payoff = variance_notional * (realized_vol**2 - variance_strike)
pd.Series({'variance_swap_payoff': payoff})
```

## 2. Mathematical Foundation

For variance notional \(N_{\text{var}}\):

$$
\text{payoff}=N_{\text{var}}\left(\sigma_{\text{realized}}^2-K_{\text{var}}\right).
$$
