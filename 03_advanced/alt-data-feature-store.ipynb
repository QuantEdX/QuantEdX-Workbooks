## 1. Introduction

An alternative-data feature store makes nontraditional signals reproducible. It records the feature definition, timestamp, source lineage, and the exact time at which a value became available to a trading model.


```python
features = pd.DataFrame({
    'asset': ['A', 'A', 'B'],
    'event_time': pd.to_datetime(['2025-01-01 09:00', '2025-01-02 09:00', '2025-01-01 09:00']),
    'available_time': pd.to_datetime(['2025-01-01 10:00', '2025-01-02 10:00', '2025-01-01 11:00']),
    'sentiment': [0.4, -0.1, 0.2],
})
prediction_time = pd.Timestamp('2025-01-01 10:30')
features[features['available_time'] <= prediction_time]
```

## 2. Mathematical Foundation

For a feature observed at event time \(t_e\) and available at processing time \(t_a\), enforce:

$$
t_a \le t_{\text{prediction}}.
$$

This point-in-time rule prevents future information from leaking into historical training data.
