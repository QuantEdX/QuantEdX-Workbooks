```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.optimize import minimize
import warnings
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline
np.random.seed(42)

print('Libraries imported successfully!')
```

## 1. Introduction

Distribution plots reveal features that averages hide: skew, heavy tails, outliers, and volatility clustering. A good first pass combines a histogram, a density estimate, quantiles, and a normal-distribution benchmark.


```python
returns = pd.Series(np.random.standard_t(df=5, size=1000) * 0.012, name='daily_return')
fig, ax = plt.subplots(figsize=(9, 4))
sns.histplot(returns, bins=40, kde=True, ax=ax)
ax.axvline(returns.quantile(0.05), color='crimson', linestyle='--', label='5% quantile')
ax.axvline(returns.mean(), color='black', linestyle=':', label='mean')
ax.legend()
pd.Series({'mean': returns.mean(), 'volatility': returns.std(), 'q05': returns.quantile(0.05), 'skew': returns.skew()})
```

## 2. Mathematical Foundation

For returns \(r_1,\ldots,r_n\), summarize location and dispersion with:

$$
\bar r = \frac{1}{n}\sum_{t=1}^{n}r_t,
\qquad
s = \sqrt{\frac{1}{n-1}\sum_{t=1}^{n}(r_t-\bar r)^2}.
$$

Quantiles such as \(q_{0.05}\) describe downside outcomes without assuming returns are normal.


## 3. Interpretation and Validation

Compare the left tail with the center of the distribution. Heavy tails make extreme moves more common than a normal approximation suggests.
