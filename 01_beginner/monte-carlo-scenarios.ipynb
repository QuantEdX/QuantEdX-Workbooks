## 1. Introduction

Monte Carlo simulation turns uncertain assumptions into a distribution of possible outcomes. In finance, simulated paths help quantify downside risk, compare strategies, and test how conclusions change under different volatility regimes.


```python
n_paths, n_days = 2000, 252
mu, sigma, dt = 0.07, 0.20, 1 / 252
shocks = np.random.normal(size=(n_days, n_paths))
log_returns = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * shocks
paths = 100 * np.exp(np.vstack([np.zeros(n_paths), log_returns.cumsum(axis=0)]))
terminal_returns = paths[-1] / paths[0] - 1
pd.Series(terminal_returns).describe(percentiles=[0.05, 0.5, 0.95])
```

## 2. Mathematical Foundation

A simple geometric Brownian motion path is:

$$
S_{t+\Delta t} = S_t
\exp\left[\left(\mu-\frac{1}{2}\sigma^2\right)\Delta t
+ \sigma\sqrt{\Delta t}Z_t\right],
\qquad Z_t \sim \mathcal{N}(0,1).
$$

For risk analysis, focus on the distribution of terminal returns and drawdowns rather than one simulated line.
