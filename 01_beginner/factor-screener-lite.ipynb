## 1. Introduction

A factor screener ranks securities using observable characteristics such as value, momentum, and quality. The goal is not to predict one exact return. It is to create a transparent, repeatable ranking process that can be tested and monitored.


```python
stocks = pd.DataFrame({
    'ticker': ['ALFA', 'BETA', 'GAMMA', 'DELTA', 'EPSI'],
    'value': [0.8, 1.4, 1.0, 1.8, 0.6],
    'momentum': [0.12, -0.03, 0.08, 0.15, 0.02],
    'quality': [0.18, 0.09, 0.14, 0.11, 0.20],
})
for column in ['value', 'momentum', 'quality']:
    stocks[f'{column}_z'] = stats.zscore(stocks[column])
stocks['score'] = stocks[['value_z', 'momentum_z', 'quality_z']].mean(axis=1)
stocks.sort_values('score', ascending=False)
```

## 2. Mathematical Foundation

For asset \(i\), standardize each factor before combining signals:

$$
z_{i,k} = \frac{x_{i,k} - \mu_k}{\sigma_k},
\qquad
\text{score}_i = \sum_{k=1}^{K} w_k z_{i,k}.
$$

The weights \(w_k\) express the importance of each factor. Use lagged inputs and rebalance at a fixed frequency to avoid look-ahead bias.


## 3. Interpretation and Validation

Inspect the ranking and ask whether a single input dominates the score. In production, winsorize extreme values, use point-in-time fundamentals, and track turnover after each rebalance.


## 4. Practical Applications

Use the screener as a research shortlist, a portfolio construction input, or a monitoring dashboard. Validate every factor out of sample before using it in a live allocation.
