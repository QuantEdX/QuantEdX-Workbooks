### What Are Options?

An **option** is a contract that gives the buyer the **right, but not the obligation**, to buy or sell an underlying asset at a specified price (strike) on or before a specified date (expiration).

**Key distinction from futures**: Options give you a choice (hence "option"), while futures create an obligation.

## 1. Core Options Terminology

### The Two Types of Options

#### Call Option
A **call option** gives the holder the **right to BUY** the underlying asset at the strike price.

- **Buyer's perspective**: Benefits when price goes UP
- **Use case**: Bullish on the underlying asset
- **Example**: Right to buy AAPL at $150

#### Put Option
A **put option** gives the holder the **right to SELL** the underlying asset at the strike price.

- **Buyer's perspective**: Benefits when price goes DOWN
- **Use case**: Bearish on the underlying asset or hedging
- **Example**: Right to sell AAPL at $150

### Key Terms

| Term | Symbol | Definition |
|------|--------|------------|
| **Underlying** | $S$ or $S_t$ | The asset the option is written on (stock, index, commodity) |
| **Strike Price** | $K$ | The price at which the option can be exercised |
| **Expiration** | $T$ | The date when the option expires (last day to exercise) |
| **Premium** | $C$ (call), $P$ (put) | The price paid to buy the option |
| **Spot Price** | $S_t$ | Current market price of the underlying |
| **Payoff** | - | Value of option at expiration (intrinsic value) |
| **Profit/Loss** | - | Payoff minus premium paid |

### Option Specifications Example

Let's look at a real option quote:

```
AAPL 150 Call 2024-12-20
```

This means:
- **Underlying**: Apple Inc. (AAPL)
- **Type**: Call option
- **Strike**: $150
- **Expiration**: December 20, 2024
- **Right**: Buy AAPL at $150 on or before Dec 20, 2024

### Long vs Short (Buyer vs Seller)

| Position | Action | Rights/Obligations | Risk | Reward |
|----------|--------|-------------------|------|--------|
| **Long Call** | Buy call | Right to buy | Limited (premium) | Unlimited |
| **Short Call** | Sell call | Obligation to sell | Unlimited | Limited (premium) |
| **Long Put** | Buy put | Right to sell | Limited (premium) | High (strike - 0) |
| **Short Put** | Sell put | Obligation to buy | High (strike - 0) | Limited (premium) |

**Key Insight**: Option buyers have **rights** with limited risk. Option sellers have **obligations** with higher risk but collect premium upfront.

## 2. Mathematical Formulas for Option Payoffs

### Call Option Payoff

At expiration, a call option's **payoff** depends on the stock price $S_T$ and strike price $K$:

$$
\text{Call Payoff} = \max(S_T - K, 0)
$$

- If $S_T > K$: Exercise the option, payoff = $S_T - K$ (in-the-money)
- If $S_T \leq K$: Don't exercise, payoff = $0$ (out-of-the-money or at-the-money)

**Profit/Loss** for long call buyer (paid premium $C$):

$$
\text{Call P\&L} = \max(S_T - K, 0) - C
$$

**Profit/Loss** for short call seller (received premium $C$):

$$
\text{Short Call P\&L} = C - \max(S_T - K, 0)
$$

### Put Option Payoff

At expiration, a put option's **payoff**:

$$
\text{Put Payoff} = \max(K - S_T, 0)
$$

- If $S_T < K$: Exercise the option, payoff = $K - S_T$ (in-the-money)
- If $S_T \geq K$: Don't exercise, payoff = $0$ (out-of-the-money or at-the-money)

**Profit/Loss** for long put buyer (paid premium $P$):

$$
\text{Put P\&L} = \max(K - S_T, 0) - P
$$

**Profit/Loss** for short put seller (received premium $P$):

$$
\text{Short Put P\&L} = P - \max(K - S_T, 0)
$$

## 3. Moneyness: ITM, ATM, OTM

**Moneyness** describes the relationship between the current stock price and the strike price.

### For Call Options

| Moneyness | Condition | Intrinsic Value | Meaning |
|-----------|-----------|-----------------|---------|
| **In-the-Money (ITM)** | $S_t > K$ | $S_t - K > 0$ | Profitable to exercise now |
| **At-the-Money (ATM)** | $S_t = K$ | $0$ | Strike equals spot price |
| **Out-of-the-Money (OTM)** | $S_t < K$ | $0$ | Would lose money if exercised |

### For Put Options

| Moneyness | Condition | Intrinsic Value | Meaning |
|-----------|-----------|-----------------|---------|
| **In-the-Money (ITM)** | $S_t < K$ | $K - S_t > 0$ | Profitable to exercise now |
| **At-the-Money (ATM)** | $S_t = K$ | $0$ | Strike equals spot price |
| **Out-of-the-Money (OTM)** | $S_t > K$ | $0$ | Would lose money if exercised |

### Example

Suppose AAPL is trading at $S_t = 155$:

- **$150 Call**: ITM (155 > 150), intrinsic value = $5
- **$155 Call**: ATM (155 = 155), intrinsic value = $0
- **$160 Call**: OTM (155 < 160), intrinsic value = $0

- **$150 Put**: OTM (155 > 150), intrinsic value = $0
- **$155 Put**: ATM (155 = 155), intrinsic value = $0
- **$160 Put**: ITM (155 < 160), intrinsic value = $5

**Key Insight**: Calls and puts have opposite moneyness at the same strike!

## 2. Mathematical Foundation

At expiration, the payoff of a long call and a long put are:

$$
\text{Call payoff} = \max(S_T-K, 0),
\qquad
\text{Put payoff} = \max(K-S_T, 0).
$$

Profit also accounts for the premium paid. For a call premium \(C_0\) and put premium \(P_0\):

$$
\text{Long call profit} = \max(S_T-K,0)-C_0,
$$

$$
\text{Long put profit} = \max(K-S_T,0)-P_0.
$$


````python
def call_payoff(stock_price, strike):
    return np.maximum(stock_price - strike, 0)

def put_payoff(stock_price, strike):
    return np.maximum(strike - stock_price, 0)

def call_profit(stock_price, strike, premium):
    return call_payoff(stock_price, strike) - premium

def put_profit(stock_price, strike, premium):
    return put_payoff(stock_price, strike) - premium
````

### Core Terminology Checklist

✅ **Call Option**: Right to BUY at strike price  
✅ **Put Option**: Right to SELL at strike price

✅ **Premium**: Price paid for the option  
✅ **Expiration**: Last date to exercise  
✅ **Long Position**: Buyer (has rights, pays premium)  
✅ **Short Position**: Seller (has obligations, receives premium)  

### Moneyness Summary

| | Call | Put |
|---|---|---|
| **ITM** | S > K | S < K |
| **ATM** | S = K | S = K |
| **OTM** | S < K | S > K |

### Payoff Formulas (At Expiration)

**Call Payoff**: $\max(S_T - K, 0)$  
**Put Payoff**: $\max(K - S_T, 0)$  
**Call P&L**: $\max(S_T - K, 0) - C$  
**Put P&L**: $\max(K - S_T, 0) - P$  

### Risk/Reward Profiles

| Position | Max Gain | Max Loss | Breakeven |
|----------|----------|----------|-----------|
| Long Call | Unlimited | Premium | K + Premium |
| Short Call | Premium | Unlimited | K + Premium |
| Long Put | K - Premium | Premium | K - Premium |
| Short Put | Premium | K - Premium | K - Premium |

### Critical Insights

1. **Options are asymmetric**: Buyers have limited downside, sellers have limited upside
2. **Zero-sum game**: Every dollar won by a buyer is lost by a seller
3. **Premium is sunk cost**: Once paid, focus on payoff at expiration
4. **Intrinsic value**: What you'd get if you exercised NOW
5. **Time value**: Premium above intrinsic value (decays to zero at expiration)
6. **Calls benefit from price increases**, **Puts benefit from price decreases**
7. **American options** can be exercised anytime, **European** only at expiration

### Common Mistakes to Avoid

1. **Confusing call and put**: Remember calls = buy right, puts = sell right
2. **Ignoring premium**: P&L must account for premium paid/received
3. **Exercising OTM options**: Never profitable (intrinsic value = 0)
4. **Selling naked options without understanding risk**: Unlimited loss potential
5. **Not knowing moneyness**: ITM/ATM/OTM determines exercise decision
6. **Confusing payoff and profit**: Payoff ignores premium, profit includes it

### Quick Reference: The Four Basic Positions

```
LONG CALL: "I think price will go UP"
- Pay premium upfront
- Max loss = premium
- Max gain = unlimited
- Use: Bullish on stock

SHORT CALL: "I think price will stay FLAT or go DOWN"  
- Receive premium upfront
- Max gain = premium
- Max loss = unlimited
- Use: Neutral to bearish, income generation

LONG PUT: "I think price will go DOWN"
- Pay premium upfront  
- Max loss = premium
- Max gain = strike price (if stock goes to $0)
- Use: Bearish on stock, portfolio protection

SHORT PUT: "I think price will stay FLAT or go UP"
- Receive premium upfront
- Max gain = premium
- Max loss = strike price (if stock goes to $0)
- Use: Neutral to bullish, acquire stock at lower price
```
