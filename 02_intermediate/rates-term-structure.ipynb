## 1. Introduction

A term structure describes discount rates across maturities. Bootstrapping extracts discount factors sequentially from market instruments so cash flows can be valued consistently.


```python
curve = pd.DataFrame({'maturity': [0.5, 1, 2, 3, 5], 'zero_rate': [0.045, 0.043, 0.041, 0.040, 0.039]})
curve['discount_factor'] = np.exp(-curve['zero_rate'] * curve['maturity'])
curve['forward_rate'] = -np.log(curve['discount_factor']).diff() / curve['maturity'].diff()
curve
```

## 2. Mathematical Foundation

For a zero-coupon discount factor \(P(0,T)\), the continuously compounded zero rate is:

$$
z(0,T) = -\frac{\ln P(0,T)}{T}.
$$

The forward rate between \(T_1\) and \(T_2\) is:

$$
f(T_1,T_2) =
\frac{\ln P(0,T_1)-\ln P(0,T_2)}{T_2-T_1}.
$$
