## 1. Introduction

News sentiment pipelines convert text into time-aligned features. The hard part is not assigning a score; it is avoiding look-ahead bias, duplicate stories, and timestamps that do not match trading availability.


```python
news = pd.DataFrame({'hours_ago': [1, 3, 8, 20], 'sentiment': [0.7, -0.2, 0.3, -0.5]})
decay = 0.12
news['weight'] = np.exp(-decay * news['hours_ago'])
signal = np.average(news['sentiment'], weights=news['weight'])
pd.Series({'decayed_sentiment_signal': signal, 'article_count': len(news)})
```

## 2. Mathematical Foundation

Aggregate article-level sentiment \(s_j\) into a decayed signal:

$$
S_t = \frac{\sum_{j: t_j \le t} e^{-\lambda(t-t_j)}s_j}
{\sum_{j: t_j \le t} e^{-\lambda(t-t_j)}}.
$$

The decay parameter \(\lambda\) controls how quickly old news loses influence.
