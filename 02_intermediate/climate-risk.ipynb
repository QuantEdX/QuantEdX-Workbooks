## 1. Introduction

Climate scenario analysis translates uncertain physical and transition risks into portfolio-level stress tests. The objective is to expose concentrations and assumptions, not to claim one precise forecast.


```python
portfolio = pd.DataFrame({'sector': ['Utilities', 'Energy', 'Technology', 'Industrials'], 'weight': [0.25, 0.20, 0.30, 0.25], 'transition_shock': [-0.08, -0.18, 0.01, -0.06]})
portfolio['contribution'] = portfolio['weight'] * portfolio['transition_shock']
pd.Series({'scenario_return': portfolio['contribution'].sum(), 'largest_loss_sector': portfolio.loc[portfolio['contribution'].idxmin(), 'sector']})
```

## 2. Mathematical Foundation

A simple scenario loss estimate is:

$$
\text{loss}_{p,s} = \sum_{i=1}^{N} w_i \times \text{shock}_{i,s}.
$$

The shocks should be scenario-specific and linked to observable exposures such as sector, region, carbon intensity, or physical location.


## 3. Interpretation and Validation

Break the scenario loss into contributions. Ask which assumptions drive the result and whether the portfolio has hidden exposure through supply chains.


## 4. Practical Applications

Use climate scenarios for risk disclosure, concentration review, and engagement priorities. Report uncertainty ranges alongside point estimates.
