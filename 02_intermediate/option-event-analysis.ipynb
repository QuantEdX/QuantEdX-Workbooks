## 1. Introduction

Event-driven option analysis compares the move implied by option prices with the move realized after earnings or macro announcements. The workflow separates volatility exposure from a directional view.


```python
events = pd.DataFrame({'event': ['A', 'B', 'C', 'D'], 'spot': [100, 80, 120, 95], 'call': [3.0, 2.2, 4.5, 2.8], 'put': [2.8, 2.4, 4.1, 2.7], 'realized_move': [0.041, 0.072, 0.035, 0.058]})
events['implied_move'] = (events['call'] + events['put']) / events['spot']
events['realized_minus_implied'] = events['realized_move'] - events['implied_move']
events
```

## 2. Mathematical Foundation

A rough at-the-money straddle approximation for the implied move is:

$$
\text{implied move} \approx \frac{C_{\text{ATM}} + P_{\text{ATM}}}{S_0}.
$$

Compare this with the absolute realized return after the event.


## 3. Interpretation and Validation

Compare implied and realized moves across events. Remember that a straddle also carries volatility, skew, and transaction-cost exposure.
