## 1. Introduction

Mean-variance optimization converts expected returns and a covariance matrix into portfolio weights. Its output is useful only when inputs, constraints, and estimation error are handled explicitly.


```python
returns = pd.DataFrame(np.random.normal(0.0004, 0.012, size=(500, 4)), columns=['Equity', 'Bonds', 'Gold', 'REIT'])
cov = returns.cov().values * 252
n_assets = cov.shape[0]
objective = lambda w: w @ cov @ w
result = minimize(objective, np.repeat(1 / n_assets, n_assets),
                  bounds=[(0, 1)] * n_assets,
                  constraints={'type': 'eq', 'fun': lambda w: w.sum() - 1})
pd.Series(result.x, index=returns.columns, name='minimum_variance_weight')
```

## 2. Mathematical Foundation

For weights \(\mathbf{w}\), expected return vector \(\boldsymbol{\mu}\), and covariance matrix \(\boldsymbol{\Sigma}\):

$$
\mathbb{E}[r_p] = \mathbf{w}^{\top}\boldsymbol{\mu},
\qquad
\sigma_p^2 = \mathbf{w}^{\top}\boldsymbol{\Sigma}\mathbf{w}.
$$

A long-only minimum-variance portfolio solves:

$$
\min_{\mathbf{w}} \ \mathbf{w}^{\top}\boldsymbol{\Sigma}\mathbf{w}
\quad \text{s.t.} \quad
\sum_i w_i=1,\; w_i\ge 0.
$$
