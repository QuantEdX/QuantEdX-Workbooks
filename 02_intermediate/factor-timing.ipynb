## 1. Introduction

Factor timing changes exposure to value, momentum, quality, or other factors as market conditions evolve. Because timing signals are noisy, the workflow should emphasize stability and out-of-sample evaluation.


```python
dates = pd.date_range('2022-01-01', periods=36, freq='ME')
signals = pd.DataFrame({'valuation': np.random.normal(size=36), 'trend': np.random.normal(size=36), 'macro': np.random.normal(size=36)}, index=dates)
signals = (signals - signals.mean()) / signals.std()
signals['timing_score'] = signals.mean(axis=1)
signals['target_tilt'] = signals['timing_score'].clip(-1, 1)
signals.tail()
```

## 2. Mathematical Foundation

A simple standardized timing score is:

$$
s_t = \sum_{k=1}^{K} w_k z_{k,t},
\qquad
\tilde w_t = \operatorname{clip}(s_t, -w_{\max}, w_{\max}).
$$

Clipping limits the effect of extreme signals and reduces turnover.


## 3. Interpretation and Validation

Check turnover, signal persistence, and performance after costs. A complicated signal with unstable exposures is unlikely to survive live trading.


## 4. Practical Applications

Use timing as a bounded tilt around a strategic allocation, not as an all-in prediction engine.
