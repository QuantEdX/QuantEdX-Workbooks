## 1. Introduction

FX carry strategies compare interest-rate differentials across currencies. Returns depend on both the yield advantage and the currency move, so attractive carry can be offset by depreciation during stressed markets.


```python
fx = pd.DataFrame({'currency': ['AUD', 'EUR', 'JPY', 'GBP'], 'rate': [0.043, 0.032, 0.002, 0.047], 'spot_return': [-0.012, 0.004, 0.018, -0.006]})
usd_rate = 0.040
fx['carry_vs_usd'] = fx['rate'] - usd_rate
fx['approx_total_return'] = fx['carry_vs_usd'] + fx['spot_return']
fx.sort_values('approx_total_return', ascending=False)
```

## 2. Mathematical Foundation

Covered interest parity links the forward rate \(F\) to spot \(S\):

$$
F_{d/f}(0,T) = S_{d/f}(0)
\exp\left[(r_d-r_f)T\right].
$$

An approximate carry return combines the rate differential with the spot move.

Separate carry income from spot return. A strategy that appears stable in calm periods may load heavily on global risk appetite.
