## 1. Introduction

Quantile risk methods model the tail of the return distribution directly. Value at Risk estimates a loss threshold, while Expected Shortfall summarizes losses beyond that threshold.


```python
returns = pd.Series(np.random.standard_t(df=5, size=3000) * 0.01)
losses = -returns
alpha = 0.95
var = losses.quantile(alpha)
expected_shortfall = losses[losses >= var].mean()
pd.Series({'VaR_95': var, 'Expected_Shortfall_95': expected_shortfall})
```

## 2. Mathematical Foundation

At confidence level \(\alpha\), define:

$$
\operatorname{VaR}_{\alpha}(L) = \inf\{\ell : \Pr(L \le \ell) \ge \alpha\},
$$

$$
\operatorname{ES}_{\alpha}(L) = \mathbb{E}[L \mid L \ge \operatorname{VaR}_{\alpha}(L)].
$$

Expected Shortfall is more informative when losses have heavy tails.


## 3. Interpretation and Validation

Compare VaR with Expected Shortfall and repeat the calculation under a more volatile regime. Tail estimates need enough observations and careful backtesting.


## 4. Practical Applications

Use tail metrics for limit setting and stress testing. Do not treat a 95% VaR estimate as the maximum possible loss.
