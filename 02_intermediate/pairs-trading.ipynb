## 1. Introduction

Pairs trading looks for two related price series whose spread tends to revert. A backtest must estimate the hedge ratio using historical information only and include trading costs.


```python
n = 500
base = np.cumsum(np.random.normal(0, 0.01, n))
prices = pd.DataFrame({'Asset_A': np.exp(base + np.random.normal(0, 0.015, n)), 'Asset_B': np.exp(base + np.random.normal(0, 0.015, n))})
spread = np.log(prices['Asset_A']) - np.log(prices['Asset_B'])
zscore = (spread - spread.rolling(60).mean()) / spread.rolling(60).std()
pd.DataFrame({'spread': spread, 'zscore': zscore}).tail()
```

## 2. Mathematical Foundation

For log prices \(p_{A,t}\) and \(p_{B,t}\), define:

$$
s_t = p_{A,t} - \beta p_{B,t},
\qquad
z_t = \frac{s_t-\mu_s}{\sigma_s}.
$$

A common rule opens a trade when \(|z_t|\) is large and closes it as the spread returns toward zero.

Plot the rolling z-score and test whether the relationship survives different time windows. Correlation alone does not establish mean reversion.
