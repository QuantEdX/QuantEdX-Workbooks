## 1. Introduction

Financial forecasting starts with a disciplined baseline. Before adding complex models, define the target, preserve time ordering, and compare against simple lagged or rolling forecasts.


```python
series = pd.Series(np.random.normal(0, 1, 500)).rolling(5).mean().dropna()
frame = pd.DataFrame({'target': series})
frame['lag_1'] = frame['target'].shift(1)
frame['rolling_mean_20'] = frame['target'].shift(1).rolling(20).mean()
frame['error_baseline'] = frame['target'] - frame['rolling_mean_20']
frame.dropna().tail()
```

## 2. Mathematical Foundation

An autoregressive model of order \(p\) is:

$$
y_t = c + \sum_{i=1}^{p}\phi_i y_{t-i} + \varepsilon_t.
$$

Evaluate forecasts on future observations only. Random train-test shuffling leaks information across time.


## 3. Interpretation and Validation

Use walk-forward validation and compare forecast error against a naive baseline. A model is useful only if its improvement persists out of sample.


## 4. Practical Applications
